```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head() #top 5 entries from dataset
data.info()#prints information about the DataFrame
pd.isnull(data).sum()#print the number of null values
```

```python
ax = sns.countplot(x = 'Sex',data = data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = data, x = 'Age', hue = 'Sex')

for bars in ax.containers:
    ax.bar_label(bars)
```
